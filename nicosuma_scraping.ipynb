```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
import pandas as pd
import re
import traceback
import random
import time
try:
    # Chrome起動
    
    
    options = Options()
    options.add_argument('--headless')  # ヘッドレスモード（画面非表示）
    options.add_argument('--window-size=1920,1080') 
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    wait = WebDriverWait(driver, 10)
    driver.get("https://www.nicosuma.com/iphone")  # ←あなたの対象ページに変更してください

  
    records = []

    # ページング
    pgae_path = "//ul[@class='m-pagination-ul']//li[last()-1]"
    page_count = driver.find_element(By.XPATH, pgae_path)
    
    print(page_count.text)
    
    for i in range(1, int(page_count.text)+1):
      print(f"{i}ページ目")
      
            # 各ページ
      iphone_list_path = "//div[@class='m-product-grid']/div"
      iphone_list = driver.find_elements(By.XPATH, iphone_list_path)
      # print("iPhoneの個数：", len(iphone_list))

      
      for iphone in iphone_list:
        WebDriverWait(driver, 10).until(EC.visibility_of(iphone))
        driver.execute_script("arguments[0].scrollIntoView();", iphone)   

        info = iphone.text
        # 中古Bグレード -> B のみを取り出す
        grade_text = info.split("\n")[0]
        grade = grade_text.replace("中古", "").replace("グレード", "")
        
        # ¥39,600 -> 39600 のみを取り出す
        price_text = info.split("\n")[4]
        price = int(price_text.replace("¥", "").replace(",", ""))

        
        
        device = info.split("\n")[2]
        # iPhone 12 64GB
        # 空白で分割
        parts = device.strip().split()
        # 後ろから1つを取り出し
        memory = parts[-1]
        # 残りをデバイス名として結合
        marketing_name = ' '.join(parts[:-1])

        # print("Grade ", grade)
        # print("Price ", price)
        # print("Marketing_name ", marketing_name)
        # print("Memory ", memory)
        # print("-------------------")
        
        records.append({
          "Marketing_name": marketing_name,
          "Memory": memory,
          "Grade": grade,
          "Price":price
        })
      
      next_page_button = driver.find_element(By.XPATH, "//ul[@class='m-pagination-ul']//li[last()]")
      next_page_button.click()
      num = random.choice([1, 2, 3, 4, 5])
      time.sleep(num)
      driver.execute_script("window.scrollTo(0, 0);")
      df = pd.DataFrame(records)    
              
except Exception as e:
    print( info )
    print("処理中にエラーが発生:", e)
    traceback.print_exc()  # 詳細なエラーログ（ファイル名・行番号・スタックトレース）を表示
    
finally:
    driver.quit()
    print("ブラウザを終了しました")
```
